## 2. relplot() - lineplot()
- 연속성 강조
- 데이터 변수의 변화: time 함수, 유사 연속 변수
- relplot(x=, y=, kind="line", data=)
  + plot이전 x 값 기준으로 데이터 정렬 > sort=False 로 비활성화 가능

```
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", kind="line", data=df);
```

```
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

### (1) 집계 및 불확실성 표현

- 큰 데이터:
  + 같은 x값에 대한 다양한 측정값이 존재하는 데이터셋 > 평균, 신뢰구간 95%로 구성해 집계

- 더 큰 데이터:
  + 신뢰구간 대신 표준편차로 구성 > 각 시점의 분포 산포

```
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```
# 신뢰구간 비활성화 (시간소요 방지)
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```
# 표준편차 plotting
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

### (2) 하위 데이터 구성(semantic mappings_의미)

> 변수 최대 3개까지 추가가능

```
# 변수 event 추가
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
# sns.relplot(x="timepoint", y="signal", hue="event", style="event",
#             kind="line", data=fmri);
```

```
# 변수 region, event 추가 (hue, style)
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
# sns.relplot(x="timepoint", y="signal", hue="region", style="event",
#             dashes=False, markers=True, kind="line", data=fmri);
```

- hue 컬럼이 수치형 데이터일 때

```
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            kind="line", data=dots);
```

```
# 특정 색상 지정
palette = sns.cubehelix_palette(start=2.8, rot=.1,light=.9, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            palette=palette,
            kind="line", data=dots);
```

```
# size > 범주형 변수도 가능 > 선의 변동성이 크면 좋지 않음
sns.relplot(x="time", y="firing_rate",
           hue="coherence", size="choice",
           palette=palette,
           kind="line", data=dots);
```

### (3) 날짜 데이터로 구성
- matplotlib의 눈금 레이블 지정 기능 활용 가능

```
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate 
# matplotlib 눈금레이블 형식
```

## 3. 여러 관계 표현하기(FacetGrid)

- col = '변수명'

## * 범주형 데이터 시각화 
- categorical plots
- catplot()
  + x, y, hue = 변수명
  + data = 데이터프레임
  + row, col = 데이터의 변수명
  + col_wrap = 숫자
  + estimator: 벡터를 스칼라(수치)로 변환
  + ci = 추정값 주변을 그리는 신뢰구간 크기
      (float, sd, None)
  + kind : 'point'/'bar'/'strip'/'swarm'/'box'/'violin'/'boxen'
  + aspect : 각 facet의 가로세로 비율
- ref. https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

## 1. 범주형 산점도
- stripplot() : default
- swarmplot()

```
# default - kind='strip'
tips = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=tips);
```

```
# 흐트러짐 방지 - jitter=False
sns.catplot(x="day", y="total_bill", jitter=False, data=tips);
```

```
# 겹침 방지 - kind='swarm'
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips);
```

```
# 의미변수 추가 - hue='변수명' (size, style 없음)
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips);
```

```
# 범주형 변수 정렬 방법(숫자일 경우 자동)
sns.catplot(x="size", y="total_bill", kind="swarm",
            data=tips.query("size != 3"))
```

```
# 범주형 변수 정렬 방법 - order=[,,]
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips);
```

```
# 세로 축으로 범주형 변수 배치
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

## 2. 범주형 분포도: 분포 정보 요약
- boxplot(): 사분위수와 5가지 요약 수치
  + 사분위수: 데이터 값들을 줄 세운 뒤 네 조각으로 나누는 세 개 지점
  + 제2사분위수(중앙값): 강건성+중심경향성
  + 데이터 값들의 절반은 Q1-Q3 범위 내
  + 최소값 / Q1 / Q2 / Q3 / 최대값
- boxenplot()
- violinplot(): KDE 플롯 + 박스플롯
  + 흰색 점: 중앙값
  + 중앙의 두꺼운 선: 사분위 범위
  + 중앙의 얇은 선: 신뢰구간(95%)


```
tips = sns.load_dataset("tips")
```

```
# 박스플롯 - kind='box'
  # 데이터의 대략적 분포와 개별적 이상치 보여줌
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```
# 변수 추가 - hue='변수명'
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips);
```

```
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.catplot(x="day", y="total_bill", hue="weekend",
            kind="box", dodge=False, data=tips);
```

```
# 박슨플롯 - kind='boxen'
  # 보다 자세한 정보 표시, 더 큰 데이터 세트에 적합
diamonds = sns.load_dataset("diamonds")
sns.catplot(x="color", y="price", kind="boxen",
            data=diamonds.sort_values("color"));
```

```
# 바이올린 플롯 - kind='violin'
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", data=tips)
```

```
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", bw=.15, cut=0,
            data=tips)
```

```
# 바이올린 분할 - split=True
  # hue 변수가 2가지로만 나뉠 때
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="violin", split=True, data=tips);
```

```
# 요약 boxplot 대신 개별값 보기 - inner='stick'
sns.catplot(x="day", y="total_bill", hue="sex",
            kind="violin", inner="stick", split=True,
            palette="pastel", data=tips);
```

```
# violinplot() + swarmplot()/stripplot()
g = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=g.ax);
```

## 3. 범주형 estimate plots
- barplot(): 전체 데이터에서의 추정값 > 신뢰국간
- countplot(): 각 범주의 관측치 수 표시
- pointplot(): 추정값, 포인트 추정치 및 신뢰구간만 표시 > hue 변수의 포인트 연결

```
# 추정치 계산 - kind='bar'
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
```

```
# 관측치 수 계산 - kind='count'
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic)
```

```
sns.catplot(y="deck", hue="class", kind="count",
            palette="pastel", edgecolor=".6",
            data=titanic)
```

```
# 포인트 추정치 - kind='point'
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
```

```
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["--", "-"],
            kind="point", data=titanic);
```

```
sns.violinplot(x=iris.species, y=iris.sepal_length)
```

```
# 플롯의 크기, 모양 제어 방법
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c");
```

## 5. Facet으로 여러 관계 보여주기

```
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips) # 각 facet의 가로세로 인치 비율
```

```
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.8,
            kind="swarm", data=tips)
```

```
# facetgrid 객체에 메서드 적용 가능
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log");
```
